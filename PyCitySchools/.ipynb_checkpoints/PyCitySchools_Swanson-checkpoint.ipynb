```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
school_path = os.path.join("Resources","schools_complete.csv")
student_path = os.path.join("Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_path)
student_data = pd.read_csv(student_path)

# Combine the data into a single dataset
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
df.head(10)
```

## District Summary

```python
schools_unique = df["school_name"].unique()
total_schools= len(schools_unique)
total_students = df["student_name"].count()
total_budget = "$" + str(school_data["budget"].sum())
average_math = round(df["math_score"].mean(), 2)
average_reading = round(df["reading_score"].mean(), 2)
overall_passing = round(((average_reading + average_math)/2), 2)
reading_above_70 = df[df["reading_score"] >= 70]
reading_passing =round(((reading_above_70["reading_score"].count())/(total_students)*100), 2)
math_above_70 = df[df["math_score"] >= 70]
math_passing = round(((math_above_70["math_score"].count())/(total_students)*100), 2)
data_for_district_summary = [{"Number of Unique Schools": total_schools, "Number of Students": total_students,\
                              "District Budget": total_budget, "Average Math Score": average_math,
                             "Average Reading Score": average_reading,"% Passing Math": math_passing,\
                             "% Passing Reading": reading_passing, "Overall Passing": overall_passing}]
pd.DataFrame(data_for_district_summary)
```

## School Summary

```python
school_summary_df = pd.DataFrame(school_data).rename(columns= {"school_name": "School Name", "type":"School Type",\
                                           "size": "School Size", "budget": "School Budget"}).sort_values(by=["School Name"]).set_index("School Name")
per_student_budget = (school_summary_df["School Budget"])/(school_summary_df["School Size"])
school_summary_df["Per Student Budget"] = per_student_budget
schools_grouped_mean = df.groupby(["school_name"]).mean()
reading_average = pd.DataFrame(schools_grouped_mean["reading_score"])
school_summary_df["Reading Score Average"] = round(reading_average,2)
math_average = pd.DataFrame(schools_grouped_mean["math_score"])
school_summary_df["Math Score Average"] = round(math_average, 2)
df_renamed = df.rename(columns={"school_name": "School Name"})
df_school= df_renamed.set_index("School Name")
df_school_passing_reading = df_school[df_school["reading_score"] >= 70]
reading_grouped = df_school_passing_reading.groupby("School Name").count()
reading_grouped_isolated = pd.DataFrame(reading_grouped["reading_score"])
percent_pass_reading = (reading_grouped_isolated["reading_score"])/(school_summary_df["School Size"])*100
per_pass_read_df = pd.DataFrame(percent_pass_reading)
school_summary_df["Percent Passing Reading"] = round(per_pass_read_df, 2)
df_school_passing_math = df_school[df_school["math_score"] >= 70]
math_grouped = df_school_passing_math.groupby("School Name").count()
math_grouped_isolated = pd.DataFrame(math_grouped["math_score"])
percent_pass_math = (math_grouped_isolated["math_score"])/(school_summary_df["School Size"])*100
per_pass_math_df = pd.DataFrame(percent_pass_math)
school_summary_df["Percent Passing Math"] = round(per_pass_math_df, 2)
overall_passing = round(((school_summary_df["Percent Passing Reading"] + school_summary_df["Percent Passing Math"])/2), 2)
school_summary_df["Overall Passing Percent"] = overall_passing
complete_school_summary_df = pd.DataFrame(school_summary_df)
```

## Scores by School Spending

```python
summdf= complete_school_summary_df
reading_grouped_isolated_df = pd.DataFrame(reading_grouped_isolated)
math_grouped_isolated_df = pd.DataFrame(math_grouped_isolated)
reading_math_df = reading_grouped_isolated_df.merge(math_grouped_isolated_df, left_index= True, right_index= True)
complete_numerical_df = summdf.merge(reading_math_df, left_index = True, right_index = True)
```

```python
spending_bins = [0, 585, 615, 645, 675]
spending_bin = ["<$585", "$585-615", "$615-645", "$645-675"]

complete_spending_reading_df = complete_numerical_df
complete_spending_math_df = complete_numerical_df

complete_spending_reading_df["Spending Range"] = pd.cut(complete_spending_reading_df["Per Student Budget"], bins=spending_bins, labels=spending_bin)
complete_spending_reading_grouped = complete_spending_reading_df.groupby("Spending Range").sum()
complete_spending_reading_passing = pd.DataFrame(round((((complete_spending_reading_grouped["reading_score"])/complete_spending_reading_grouped["School Size"])*100),2))

complete_spending_math_df["Spending Range"] = pd.cut(complete_spending_math_df["Per Student Budget"], bins=spending_bins, labels=spending_bin)
complete_spending_math_grouped = complete_spending_math_df.groupby("Spending Range").sum()
complete_spending_math_passing = pd.DataFrame(round((((complete_spending_math_grouped["math_score"])/complete_spending_math_grouped["School Size"])*100),2))

complete_spending_math_reading_passing = complete_spending_reading_passing.merge(complete_spending_math_passing, left_index = True, right_index = True)
complete_spending_math_reading_passing_renamed = complete_spending_math_reading_passing.rename(columns={"0_x": "Percent Passing Reading", "0_y": "Percent Passing Math"})
overall_passing_spending = round(((complete_spending_math_reading_passing_renamed["Percent Passing Reading"] + complete_spending_math_reading_passing_renamed["Percent Passing Math"])/2), 2)
complete_spending_math_reading_passing_renamed["Overall Passing Percent"] = overall_passing_spending
```

```python
spendingdf = complete_school_summary_df
spendingdf["Spending Range"] = pd.cut(spendingdf["Per Student Budget"], bins=spending_bins, labels=spending_bin)
spending_df_groupby = round((spendingdf.groupby("Spending Range").mean()), 2)
spending_df_groupby_mean = spending_df_groupby.drop(columns= ["School ID", "School Size", "School Budget",\
                                                             "Per Student Budget", "Percent Passing Reading",\
                                                             "Percent Passing Math", "Overall Passing Percent"])
round(spending_df_groupby_mean,2)
final_school_spending_df = spending_df_groupby_mean.merge(complete_spending_math_reading_passing_renamed, left_index = True, right_index = True)
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
group_names_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
complete_size_reading_df = complete_numerical_df
complete_size_math_df = complete_numerical_df

complete_size_reading_df["Size Range"] = pd.cut(complete_size_reading_df["School Size"], bins=size_bins, labels=group_names_size)
complete_size_reading_grouped = complete_size_reading_df.groupby("Size Range").sum()
complete_size_reading_passing = pd.DataFrame(round((((complete_size_reading_grouped["reading_score"])/complete_size_reading_grouped["School Size"])*100),2))

complete_size_math_df["Size Range"] = pd.cut(complete_size_math_df["School Size"], bins=size_bins, labels=group_names_size)
complete_size_math_grouped = complete_size_math_df.groupby("Size Range").sum()
complete_size_math_passing = pd.DataFrame(round((((complete_size_math_grouped["math_score"])/complete_size_math_grouped["School Size"])*100),2))

complete_size_math_reading_passing = complete_size_reading_passing.merge(complete_size_math_passing, left_index = True, right_index = True)
complete_size_math_reading_passing_renamed = complete_size_math_reading_passing.rename(columns={"0_x": "Percent Passing Reading", "0_y": "Percent Passing Math"})
overall_passing_size = round(((complete_size_math_reading_passing_renamed["Percent Passing Reading"] + complete_size_math_reading_passing_renamed["Percent Passing Math"])/2), 2)
complete_size_math_reading_passing_renamed["Overall Passing Percent"] = overall_passing_size
```

```python
sizedf = complete_school_summary_df
sizedf["Size Range"] = pd.cut(sizedf["School Size"], bins=size_bins, labels=group_names_size)
size_df_groupby = round((sizedf.groupby("Size Range").mean()), 2)
size_df_groupby_mean = size_df_groupby.drop(columns= ["School ID", "School Size", "School Budget",\
                                                             "Per Student Budget", "Percent Passing Reading",\
                                                             "Percent Passing Math", "Overall Passing Percent"])
round(size_df_groupby_mean,2)
final_school_size_df = size_df_groupby_mean.merge(complete_size_math_reading_passing_renamed, left_index = True, right_index = True)
```
